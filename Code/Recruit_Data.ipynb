```python
# Get recurits for Florida from 2000 to 2025 and store in a dataframe
df = pd.DataFrame()
for i in range(2000, 2026):
    recruits = api.get_recruits(year=i, team="Florida")
    recruit_list = []
    for r in recruits:
        recruit_list.append({
            "id": r.id,
            "athlete_id": r.athlete_id,
            "name": r.name,
            "school": r.school,
            "committed_to": r.committed_to,
            "position": r.position,
            "ranking": r.ranking,
            "stars": r.stars,
            "rating": r.rating,
            "city": r.city,
            "state_province": r.state_province,
            "year": r.year,
            "recruit_type": r.recruit_type
        })
    df = pd.concat([df, pd.DataFrame(recruit_list)], ignore_index=True)
```
