```python
# Make list of sec schools to loop through
schools = ['Florida', 'Georgia', 'Alabama', 'LSU', 'Auburn', 'Mississippi State', 'Ole Miss', 'Arkansas', 'Missouri', 'South Carolina', 'Texas A&M', 'Vanderbilt', 'Tennessee', 'Kentucky', 'Texas', 'Oklahoma']
```

### NOTE: API has limitations per call and with this list only 2 years is allowed at a time. Hypotheically, 20 year of data requires 10 code cells. However, certain 2 year ranges had more recruits in them then others and require their own cells. These include: 07-08, 11-12, 15-16, 19-20, 23-24.

```python
# Get recurits for SEC schools from 2005 to 2006 and store in a dataframe
df_05_06 = pd.DataFrame() # make empty dataframe to store data in
for i in range(2005, 2007): # loop through years 2005 to 2006
    for school in schools:
        recruits = api.get_recruits(year=i, team=school)
        recruit_list = []
        for r in recruits:
            recruit_list.append({
                "id": r.id,
                "athlete_id": r.athlete_id,
                "name": r.name,
                "school": r.school,
                "committed_to": r.committed_to,
                "position": r.position,
                "ranking": r.ranking,
                "stars": r.stars,
                "rating": r.rating,
                "city": r.city,
                "state_province": r.state_province,
                "year": r.year,
                "recruit_type": r.recruit_type
            })
        df_05_06 = pd.concat([df_05_06, pd.DataFrame(recruit_list)], ignore_index=True)
```
